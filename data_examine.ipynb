```python
import pandas as pd
import matplotlib as plt

infiles = ['./data/MA-sequences-all.csv',
            './data/TX-sequences-all.csv',
            './data/CA-sequences-all.csv',
            './data/FL-sequences-all.csv',
            './data/MD-sequences-all.csv',
            './data/IL-sequences-all.csv',
            './data/GA-sequences-all.csv',
            './data/CT-sequences-all.csv',
            './data/AZ-sequences-all.csv',
            './data/AL-sequences-all.csv',
          ]

# infiles = ['./data/AL-sequences-all.csv']
for idx, infile in enumerate(infiles):    
    print('Reading csv file {}'.format(infile))
    df = pd.read_csv(infile)
    
    if idx == 0:
        df_all = df
    else:
        df_all = pd.concat([df_all, df])

print('Sorting')
df_all = df_all.sort_index(axis=0, level=['Dates', 'C->T', 'T->C', 'C->A'])

df_all.iloc[:, 3:].plot(figsize=(12, 12), subplots=True)

outfile = './data/USA-sorted.csv'
print('Writing csv file {}'.format(outfile))

df_all.to_csv(outfile, index=False)
```
