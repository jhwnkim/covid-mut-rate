```python
import pandas as pd
import matplotlib as plt

# infiles = ['./data/MA-sequences-all.csv',
#             './data/TX-sequences-all.csv',
#             './data/CA-sequences-all.csv',
#             './data/FL-sequences-all.csv',
#             './data/MD-sequences-all.csv',
#             './data/IL-sequences-all.csv',
#             './data/GA-sequences-all.csv',
#             './data/CT-sequences-all.csv',
#             './data/AZ-sequences-all.csv',
#             './data/AL-sequences-all.csv',
#            './data/MN-sequences-all.csv',
#            './data/NC-sequences-all.csv',
#            './data/NJ-sequences-all.csv',
#            './data/NM-sequences-all.csv',
#            './data/NY-sequences-all.csv',
#            './data/OH-sequences-all.csv',
#            './data/PA-sequences-all.csv',
#            './data/RI-sequences-all.csv',
#            './data/SC-sequences-all.csv',
#            './data/TN-sequences-all.csv',
#            './data/VA-sequences-all.csv',
#            #'./data/WA-sequences-all.csv',
#            './data/WI-sequences-all.csv',
#           ]
infiles = ['./data/MA-sequences-all-avg.csv',
            './data/TX-sequences-all-avg.csv',
            './data/CA-sequences-all-avg.csv',
            './data/FL-sequences-all-avg.csv',
            './data/MD-sequences-all-avg.csv',
            './data/GA-sequences-all-avg.csv',
            './data/CT-sequences-all-avg.csv',
            './data/AZ-sequences-all-avg.csv',
            './data/AL-sequences-all-avg.csv',
           './data/MN-sequences-all-avg.csv',
           './data/NC-sequences-all-avg.csv',
           './data/NJ-sequences-all-avg.csv',
           './data/NM-sequences-all-avg.csv',
           './data/NY-sequences-all-avg.csv',
           './data/OH-sequences-all-avg.csv',
           './data/PA-sequences-all-avg.csv',
           './data/RI-sequences-all-avg.csv',
           './data/SC-sequences-all-avg.csv',
           './data/TN-sequences-all-avg.csv',
           './data/VA-sequences-all-avg.csv',
           './data/WA-sequences-all-avg.csv',
           './data/WI-sequences-all-avg.csv',
          ]

# infiles = ['./data/AL-sequences-all.csv']
for idx, infile in enumerate(infiles):    
    print('Reading csv file {}'.format(infile))
    df = pd.read_csv(infile)
    
    if idx == 0:
        df_all = df
    else:
        df_all = pd.concat([df_all, df])
        
df_raw = df_all
print('Removing entries with insufficient date information')
df_all = df_raw[df_raw['Dates'].map(len)>7]

print('Sorting')
df_all.sort_values(by=['Dates', 'A->C', 'A->T', 'A->G', 'C->A', \
                                   'C->T', 'C->G', 'T->A', 'T->C', 'T->G', \
                                   'G->A', 'G->C', 'G->T'], axis=0, inplace=True, ignore_index=True)

print(df_all.head(10))

df_all.iloc[:, 3:].plot(figsize=(12, 12), subplots=True)

outfile = './data/USA-sorted.csv'
print('Writing csv file {}'.format(outfile))

df_all.to_csv(outfile, index=False)
```
